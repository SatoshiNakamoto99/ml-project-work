```python
import pandas as pd
base_model = "mobilenet-v3s-more-params-multi"#"mobilenet-v3l-more-params"#"YOLOv8n-cls-bin"
training_dataset = "foggia-unbalanced-ext"#"foggia-unbalanced-ext"
models_root = "experiments"
model_name = "mobilenet-v3s-more-params-multi"#"YOLOv8n-cls-bin"experiments/YOLOv8n-cls-bin/foggia-unbalanced
lstm_mode = False
multilabel = True

for k in range(10):
    model_path = f"fold_{k}_best_model.pth"

    print(f"videos path: datasets/{training_dataset}/VIDEOS/TEST_SET/")
    print(f"results path: results/{model_name}/{training_dataset}/")
    !python3 test.py --videos datasets/{training_dataset}/VIDEOS/TEST_SET/0/ --results results/{model_name}/{training_dataset}/0/ --base_model {base_model} --training_dataset {training_dataset} --models_root {models_root} --model_name {model_name} --model_path {model_path} {"--lstm_mode True" if lstm_mode else ""} {"--multilabel True" if multilabel else ""}
    !python3 test.py --videos datasets/{training_dataset}/VIDEOS/TEST_SET/1/ --results results/{model_name}/{training_dataset}/1/ --base_model {base_model} --training_dataset {training_dataset} --models_root {models_root} --model_name {model_name} --model_path {model_path} {"--lstm_mode True" if lstm_mode else ""} {"--multilabel True" if multilabel else ""}
    print(f"predictions path: results/{model_name}/{training_dataset}/")
    print(f"gt path: datasets/{training_dataset}/GT/TEST_SET/")
    !python3 eval.py --predictions results/{model_name}/{training_dataset}/ --gt datasets/{training_dataset}/GT/TEST_SET/ --fold {k}

#foreach fold
for k in range(10):
    #read the results
    df = pd.read_csv(f"results/{model_name}/{training_dataset}/fold_{k}_results.csv")
    #and concat to single csv
    if k == 0:
        df_all = df
    else:
        df_all = pd.concat([df_all, df])
#save the csv
df_all.to_csv(f"results/{model_name}/{training_dataset}/results.csv", index=False)
```

```python
import pandas as pd
import os
import math
import glob
#experiment_name = "experiments/mobilenet-v3l/foggia-mod/"
#experiment_name = "experiments/resnet50-test/foggia-mod/"

def print_kfold_results(experiment):
    best_models_metrics_path = os.path.join(experiment, "best_models.csv")
    df = pd.read_csv(best_models_metrics_path)
    # print(f"Experiment {experiment} results:")
    # print("\tMean val_acc:", df["val_acc"].mean())
    # print("\tStandard deviation val_acc:", df["val_acc"].std())
    # print("\tStandard error:", df["val_acc"].std()/math.sqrt(10))
    # print("\tWith 0.95 probability the mean is in the interval:", df["val_acc"].mean()-2*df["val_acc"].std()/math.sqrt(10), df["val_acc"].mean()+2*df["val_acc"].std()/math.sqrt(10))
    return df

df_all_train = pd.DataFrame()
df_all_test = pd.DataFrame()

for experiment in glob.glob("experiments/*/*"):
    if not os.path.exists(os.path.join(experiment, "best_models.csv")):
        continue
    df = print_kfold_results(experiment)
    val_acc_stats = df.aggregate({"val_acc":['mean', 'std']}).transpose().set_index(pd.Index([experiment]))
    df_all_train = pd.concat([df_all_train, val_acc_stats])
    
    inference_data_path = os.path.join("results",experiment.split("/",1)[1], "results.csv")
    
    if os.path.exists(inference_data_path):
        df_inference = pd.read_csv(inference_data_path)
        inference_stats = df_inference.aggregate({"accuracy":['mean', 'std']}).transpose().set_index(pd.Index([experiment]))
        inference_stats = inference_stats.rename(columns={"mean": "test_mean", "std": "test_std"})
        df_all_test = pd.concat([df_all_test, inference_stats])

df_all = df_all_train.join(df_all_test)
df_all["Augmentation"] = df_all.apply(lambda x: "aug" in x.name, axis=1)
df_all["Extended dataset"] = df_all.apply(lambda x: "ext" in x.name, axis=1)
df_all["Stratified"] = df_all.apply(lambda x: "stratified" in x.name, axis=1)
df_all["Multilabel"] = df_all.apply(lambda x: "multi" in x.name, axis=1)
print("All experiments results:")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    pd.set_option('display.max_colwidth', None)
    display(df_all.sort_values(by=["test_mean"], ascending=False))
```

```python
import torch
import torch.utils.benchmark as benchmark
from models.mobilenet_v3_large import mobilenet_v3l
from models.mobilenet_v3_small import mobilenet_v3s
from models.resnet import resnet_sharma
from ultralytics import YOLO
yolo = YOLO("./models/saved/YOLOv8n_cls/yolov8n-cls.pt")
torch.cuda.reset_peak_memory_stats()
model = mobilenet_v3s(2).to('cuda')
#model = yolo.model.to('cuda')
#model = resnet_sharma(2).to('cuda')
def get_model_fps(model, inputs):
    

    # Esegue il benchmark del modello
    benchmark_results = benchmark.Timer(
        stmt='model(inputs)',
        setup='model.eval()',
        globals={'model': model, 'inputs': inputs}
    ).blocked_autorange(min_run_time=1)

    # Calcola gli FPS del modello
    fps = 1 / benchmark_results.median
    return fps
input_shape = (3, 224, 224)
# Crea un input di esempio
inputs = torch.randn(1, *input_shape).to('cuda')
fps = get_model_fps(model, inputs)
print(f"FPS: {fps:.2f}")
print("Max memory allocated: {} GB".format(torch.cuda.max_memory_allocated() / (1024 * 1024 * 1024)))
```
